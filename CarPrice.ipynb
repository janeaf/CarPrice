## **Correlación Específica**
Vamos a elegir una columna y especificaremos un método.

```python
def correlation(dataframe,threshold):
  cols_corres = set()
  corr_matrix = dataframe.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) >= threshold:
                 col_name = corr_matrix.columns[i]
                 cols_corr.add(col_name)
  return cols_corres
```

## **Separar por Categórias**

No númerico.

```python
def categorical(dataframe):
  categorical_col = []
  for col in dataframe:
    if dataframe[col].dtype == 'object':
      categorical_col.append(col)
  return categorical_col
```

```python
categorical_col = categorical(dataframe)
print(categorical_col)
```

## **Separar por Categórias**

Númerico.

```python
def numerical(dataframe):
  numerical_col = []
  for col in dataframe:
    if dataframe[col].dtype != 'object':
      numerical_col.append(col)
  return numerical_col
```

```python
numerical_col = numerical(dataframe)
print(numercal_col)
```

## **Una sola columna **

```python
i = 1
plt.figure(figsize=(30,120))
for col in categorical_col:
  plt.subplot(len(categorical_col),2,i)
  sns.countplot(dataframe[col])

sns.countplot(dataframe['fueltype'])
```

## **Todas las columnas**

```python
i=1
plt.figure(figsize=(30,120))
for col in categorical_col:
  plt.subplot(len(categorical_col),2,i)
  sns.countplot(dataframe[col])
  plt.yticks(fontsize=20)
  plt.xticks(fontsize=20)
  plt.xlabel(col,fontsize=20)
  plt.ylabel("count",fontsize=20)
  i = i + 1
```

## **Categórias en Columnas con COUNT y BOX PLOT**

```python
i=1
plt.figure(figsize=(30,120))
for col in categorical_col:
  # COUNT PLOT
  plt.subplot(len(categorical_col),2,i)
  sns.countplot(dataframe[col])
  plt.yticks(fontsize=20)
  plt.xticks(fontsize=20)
  plt.xlabel(col,fontsize=20)
  plt.ylabel("count",fontsize=20)

  i = i+1

  # BOX PLOT
  plt.subplot(len(categorical_col),2,i)
  sns.boxplot(x=dataframe[col], y=dataframe['stroke'])
  plt.yticks(fontsize=20)
  plt.xticks(fontsize=20)
  plt.xlabel(col,fontsize=20)
  plt.ylabel("count",fontsize=20)

  i = i+1
plt.show()
```
